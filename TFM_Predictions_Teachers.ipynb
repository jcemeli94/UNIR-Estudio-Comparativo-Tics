# Importación de librerias necesarias

```python
#Para esta actividad se importarán las siguientes librerías:
import matplotlib.pyplot as plt
import pandas as pd

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping

####### Preprocesamiento
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

####### Regresión lineal
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

```

```python
#Cargamos el modelo
model = keras.models.load_model("CacheTeachers/teachers.keras")
```

```python
#Cargamos las predicciones
df_predictions_spain = pd.read_csv("TeacherExamplePredictions/Teachers_Spain_Median.csv", sep=";")
df_predictions_spain_no_att = pd.read_csv("TeacherExamplePredictions/Teachers_Spain_Median_Att_Disabled.csv", sep=";")
df_predictions_spain_en_att = pd.read_csv("TeacherExamplePredictions/Teachers_Spain_Median_Att_Enabled.csv", sep=";")
df_predictions_spain_li_no_att = pd.read_csv("TeacherExamplePredictions/Teachers_Spain_Median_li_Att_Disabled.csv", sep=";")
df_predictions_spain_li_en_att = pd.read_csv("TeacherExamplePredictions/Teachers_Spain_Median_li_Att_Enabled.csv", sep=";")

#Retiramos los datos PISA para ajustar el dataframe al modelo de predicción
df_predictions_spain_noPISA = df_predictions_spain.drop(columns=["PISA_Media"])
df_predictions_spain_no_att_noPISA = df_predictions_spain_no_att.drop(columns=["PISA_Media"])
df_predictions_spain_en_att_noPISA = df_predictions_spain_en_att.drop(columns=["PISA_Media"])
df_predictions_spain_li_no_att_noPISA = df_predictions_spain_li_no_att.drop(columns=["PISA_Media"])
df_predictions_spain_li_en_att_noPISA = df_predictions_spain_li_en_att.drop(columns=["PISA_Media"])

#Realizamos las normalizaciones:
loaded_stats = pd.read_csv('CacheTeachers/train_stats.csv', index_col=0)

def norm(x):
  return (x - loaded_stats['mean']) / (loaded_stats['std'] - 1e-9)
    
#Realizamos las predicciones
pred_spain = model.predict(norm(df_predictions_spain_noPISA))
pred_spain_no_att = model.predict(norm(df_predictions_spain_no_att_noPISA))
pred_spain_en_att = model.predict(norm(df_predictions_spain_en_att_noPISA))
pred_spain_li_no_att = model.predict(norm(df_predictions_spain_li_no_att_noPISA))
pred_spain_li_en_att = model.predict(norm(df_predictions_spain_li_en_att_noPISA))

#Display
print("Predicción España:", pred_spain[0][0], " PISA Real: ", df_predictions_spain["PISA_Media"][0])
print("----------------------------------------------------------------------------------------")
print("Predicción España disabled att:", pred_spain_no_att[0][0], " PISA predict original: ", pred_spain[0][0])
print("Predicción España enabled att:", pred_spain_en_att[0][0], " PISA predict original: ", pred_spain[0][0])



print("Predicción España:", pred_spain[0][0], " PISA Real: ", df_predictions_spain["PISA_Media"][0])
print("----------------------------------------------------------------------------------------")
print("Predicción España less importance disabled att:", pred_spain_li_no_att[0][0], " PISA predict original: ", pred_spain[0][0])
print("Predicción España less importance enabled att:", pred_spain_li_en_att[0][0], " PISA predict original: ", pred_spain[0][0])
```
